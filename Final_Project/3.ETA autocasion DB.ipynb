```python
def class_fuel(dataframe):
    conditions = [dataframe['Fuel'].str.match('Diesel'),
                  dataframe['Fuel'].str.match('Gas'),
                  dataframe['Fuel'].str.match('Diésel'),
                  dataframe['Fuel'].str.match('Gasolina/gas'),
                  dataframe['Fuel'].str.match('Corriente eléctrica'),
                  dataframe['Fuel'].str.match('Gasolina'),
                  dataframe['Fuel'].str.match('Híbrido'),
                  dataframe['Fuel'].str.match('Gasolina y corriente eléctrica'),
                  dataframe['Fuel'].str.match('Diesel y corriente eléctrica'),
                  dataframe['Fuel'].str.match('Eléctrico')]

    choices =['Diesel','Gasoline','Diesel','Gasoline','Electric','Gasoline','Hybrid','Hybrid','Hybrid','Electric']

    dataframe['Fuel'] = np.select(conditions,choices)

    return dataframe
```

```python
def class_motor(dataframe):
    conditions = [dataframe['Motor'].str.match('Automático'),
                  dataframe['Motor'].str.match('Manual')]

    choices =['Automatic','Manual']

    dataframe['Motor'] = np.select(conditions,choices)

    return dataframe
```

```python
#Convert column to only numbers and to numeric format
def get_numbers(column):
    main_car_database[column] = main_car_database[column].apply(lambda x: re.sub(r'[^0-9]','', str(x)))
    main_car_database[column] = pd.to_numeric(main_car_database[column])
    return column
```

```python
#Eliminate specific text ina particular column
def clean_text_to_num(column,text):
    
    main_car_database[column] = main_car_database[column].str.replace(text,'', regex=False)
    
    return column
```

```python
def class_chasis(dataframe):
    conditions = [dataframe['Chasis'].str.match('Berlina'),
                  dataframe['Chasis'].str.match('Berlina mediana o grande'),
                  dataframe['Chasis'].str.match('Familiar'),
                  dataframe['Chasis'].str.match('Pequeño'),
                  dataframe['Chasis'].str.match('Monovolumen'),
                  dataframe['Chasis'].str.match('4x4, SUV o pickup'),
                  dataframe['Chasis'].str.match('Todo Terreno'),
                  dataframe['Chasis'].str.match('Stationwagon'),
                  dataframe['Chasis'].str.match('Deportivo o coupé'),
                  dataframe['Chasis'].str.match('Not defined'),
                  dataframe['Chasis'].str.match('Coupe'),
                  dataframe['Chasis'].str.match('Convertible'),
                 dataframe['Chasis'].str.match('Descapotable o convertible'),
                 ]

    choices =['Berlina','Berlina','Familiar','Pequeño','Monovolumen','SUV','Todo Terreno','Stationwagon',
              'Coupe','Not defined','Coupe','Convertible','Convertible']

    dataframe['Chasis'] = np.select(conditions,choices)

    return dataframe
```

```python
#Lets classify the colors into less optiopms
def class_color(dataframe):
    conditions = [dataframe['Color'].str.match('blanco'),
                  dataframe['Color'].str.match('gris'),
                  dataframe['Color'].str.match('negro'),
                  dataframe['Color'].str.match('azul'),
                  dataframe['Color'].str.match('plata'),
                  dataframe['Color'].str.match('gris plata'),
                  dataframe['Color'].str.match('marrón'),
                  dataframe['Color'].str.match('beige'),
                  dataframe['Color'].str.match('gris / metalizado'),
                  dataframe['Color'].str.match('blanco / metalizado'),
                  dataframe['Color'].str.match('plateado'),
                  dataframe['Color'].str.match('negro / metalizado'),
                  dataframe['Color'].str.match('naranja'),
                  dataframe['Color'].str.match('granate'),
                  dataframe['Color'].str.match('verde'),
                  dataframe['Color'].str.match('azul / metalizado')]

    choices =['blanco','gris','negro','azul','plata','gris plata','marrón','beige',
              'gris / metalizado','blanco / metalizado','plateado','negro / metalizado',
              'naranja','granate','verde','azul / metalizado']

    dataframe['Color'] = np.select(conditions,choices,default='Other')

    return dataframe
```

```python
def class_segment(dataframe):
    conditions = [(dataframe['Length_mm'] <= 3700),
                  (dataframe['Length_mm'] > 3700) & (dataframe['Length_mm'] <= 4000 ),
                  (dataframe['Length_mm'] > 4000) & (dataframe['Length_mm'] <= 4500 ),
                  (dataframe['Length_mm'] > 4500) & (dataframe['Length_mm'] <= 4700 ),                 
                  (dataframe['Length_mm'] > 4700) & (dataframe['Length_mm'] <= 5000 ),
                  (dataframe['Length_mm'] > 5000)]                
                  

    choices =['SegmentA','SegmentB','SegmentC','SegmentD','SegmentE','SegmentF']

    dataframe['Segment'] = np.select(conditions,choices)

    return dataframe
```
